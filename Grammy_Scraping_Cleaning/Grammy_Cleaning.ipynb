# Verunreinigungen aus dem gescrapten DataFrame beseitigen

### Verunreinigte csv-Datei laden

```python
grammy_relevant_dirty = pd.read_csv("Suter_Sandro_dirty.csv")
grammy_relevant_dirty
```

### Nicht notwendige Spalten entfernen

Beim Scraping wurde zuerst für jedes Jahr ein DataFrame mit den relevanten Kategorien erstellt.

Da beim zusammenführen zu dem finalen DataFrame (grammy_relevant.csv) wurde der concat-Befehl aus der Bibliothek "Pandas" verwendet. Dieser fügt die DataFrames über die sogenannten Keys zusammen. Als Keys wurden die relevanten Kategorien gewählt.

Dies führte dazu, dass die Grammy-Kategorien 2x im finalen DataFrame auftauchen. Weiter wurde auch die Index-Nummer der einzelnen DataFrames im finalen DataFrame abgebildet. Somit sind die Spalten "category" und "Unnamed: 2" nicht mehr von nöten und können gelöscht werden.

Dies geschieht über die _.drop()_-Methode des Pandas-DataFrames. Mit _"columns='...' "_ wird die Spalte gewählt, welche gelöscht werden soll. Durch den Befehl _"inplace=True"_ wird angegeben, dass die Methode direkt auf den Originaldaten ausgeführt werden soll.

```python
grammy_relevant_dirty.drop(columns='Unnamed: 2', inplace=True)
grammy_relevant_dirty.drop(columns='category', inplace=True)
grammy_relevant_dirty
```

### Benötigte Spalten bezeichnen

Durch das Zusammenführen der verschiedenen DataFrames je Jahr sind Spalten ohne Namen entstanden. Diese müssen nun benennt werden.

Dies wird über die _.rename()_-Methode gemacht. Durch die Übergabe eines Dictionaries, in welchem der _Key= 'Bestehender Spaltenname'_ und der _Value= 'Gewünschter Spaltenname'_ ist, können die Spaltennamen angepasst werden. Auch hier wird wieder mit der Einstellung _inplace=True_ gearbeitet, damit direkt die Originaldaten angepasst werden. 

```python
grammy_relevant_dirty.rename(columns={'Unnamed: 0': 'Kategorie', 'Unnamed: 1': 'Jahr', 'Song titel': 'Winner','artist':'Künstler/in'}, inplace=True)
```

```python
grammy_relevant_dirty
```

### Jahreszahlen von 20019 zu 2019 wechseln

Die von Hand eingeführte Verunreinigung des Jahres 2019 zu _20019_ wird mit der _.replace()_-Methode behoben. Die Spalte _'Jahr'_ wird dabei auf den Ausdruck _20019_ untersucht und dieser wird immer durch _2019_ ersetzt. Auch diese Anpassung geschieht inplace.

```python
grammy_relevant_dirty['Jahr'].replace(20019,2019, inplace=True)
grammy_relevant_dirty
```

### Duplikat entfernen

Das am Ende des DataFrames eingefügte Duplikat muss nun mittels einer Methode entfernt werden. Bei der Durchsicht des DataFrames kann festgestellt werden, dass am Schluss die Index-Nummer 50 nochmals vorkommt. Dies lässt vermuten, dass es sich um ein Duplikat handelt. 

Daher werden zuerst die Indexnummern 50 und 118 angeschaut, Ziel ist es zu vergleichen ob es ein Dupplikat ist:

```python
print(grammy_relevant_dirty.iloc[50])
print("*"*10)
print(grammy_relevant_dirty.iloc[118])
```

Es lässt sich unschwer erkennen, dass es sich um ein Duplikat handelt, welches entfernt werden muss (beide sind Album Of The Year, 2012).

Dazu wird die Methode _.drop_duplicates()_ verwendet. Diese geht das gesamte DataFrame durch und entfernt allfällige Duplikate:

```python
grammy_relevant_dirty = grammy_relevant_dirty.drop_duplicates()
grammy_relevant_dirty
```

### Entfernen von "()" aus der Spalte "Künstler/in"

In der Spalte "Künstler/in" gibt es mehrere Einträge, welche fälschlicherweise in einer Klammer geschrieben werden. Als Beispiel wird hier die Index-Nr. 115 gezeigt:

*Best Long Form Music Video	2003	Legend	**(Sam Cooke)**	,No Interpreten,*

Sam Cooke sollte hier nicht in den Klammern stehen. Besonders zu beachten ist, dass es Zeilen gibt, bei denen die Klammern Sinn ergeben. Bei diesen sollen die Klammern weiterhin bestehen. Als Beispiel wird die Index-Nr. 116 genommen:

*Best Long Form Music Video	2001	Recording The Producers - A Musical Romp With ...	**Mel Brooks (Various Artists Including Nathan L...**	,No Interpreten,*

Hierbei ist der Text innerhalb der Klammer eine Anmerkung und soll so bestehen bleiben.

Um ein Überblick zu bekommen, wo dass es alles Klammern gibt, die entfernt werden müssen und wo es solche hat, die bestehen bleiben sollen, wird zuerst die _.unique()_-Methode angewandt. Diese zeigt sämtliche Werte, die in der Spalte _'Künstler/in'_ vorkommt.

```python
print(grammy_relevant_dirty['Künstler/in'].unique())
```

Es lässt sich schnell erkennen, dass obiges Beispiel (Zeile 116) die einzige Zeile ist, die berechtigte Klammern hat. Somit müssen sämtliche Klammern aus der Spalte _'Künstler/in'_ entfernt werden, bis auf diejenigen der Zeile 116.

Zuerst werden sämtliche Klammern am Anfang und am Ende jedes Strings mittels regex entfernt. Dazu wird mit der Methode _.loc[]_ die Spalte _'Künstler/in'_ aufgerufen und mit der Methode _.replace()_ die Klammern durch _''_ (nichts) ersetzt.

Innerhalb der _replace()_-Methode wird ein Regex angewendet. Dies wird dann auch mittels _regex=True_ definiert. Der angewendete Regex-Befehl ist:

**^\(|)$**

Durch das Zeichen **^** wird zuerst am Anfang des Strings nach **(** gesucht. Durch das Zeichen **|** wird dann zur nächsten Expression gesprungen. Diese sucht nach **)** am Schluss des Stings.

Zusammenfassend wird durch den Regex **^\(|\)$** die erste und letzte Klammer eines Stings ausgegeben. Beispiel:

**(**Sam Cook**)**

Mel Brooks (Various Artists Including Nathan Lane & Matthew Broderick**)**

**(**John Lenon**)**

```python
grammy_relevant_dirty.loc[:'Künstler/in'].replace('^\(|\)$', '', regex=True, inplace=True)
```

Auf der Zeile 116 ist nun eine Klammer zu viel entfernt. Daher wird dieser Eintrag wieder von Hand korrigiert. Auch hier wird mittels der _.loc_-Methode auf den Eintrag zugegeriffen und dieser mittels _=_ manuell angepasst.

```python
grammy_relevant_dirty.loc[116,'Künstler/in']='Mel Brooks (Various Artists Including Nathan Lane & Matthew Broderick)'
```

```python
grammy_relevant_dirty['Künstler/in'][116]
```

### Spalten "artist2" & "Künstler/in" zu "_Künstler/in umfassend" zusammenlegen

Die Spalte "artist2" wurde beim Scraping aufgrund der Beschaffenheit der Webseite der Grammy-Awards erstellt. Die Namen der Künstlerinnen und Künstler, welche jeweils gewonnen haben, waren dort nicht immer im gleichen HTML-Element.

Nun gilt es eine neue Spalte _'_Künstler/in umfassend'_ zu erstellen, welche keine NaN Werte mehr enthält und jeweils den oder die Künstlerin enthaltet, welche die Kategorie im entsprechenden Jahr gewonnen hat.

Dazu wird zuerst geschaut wie viele NaN es hat:

```python
print("Anzahl NaN in 'Künstler/in' vorher: ", grammy_relevant_dirty['Künstler/in'].isna().sum())
print("Anzahl NaN in 'artist2' vorher: ", grammy_relevant_dirty['artist2'].isna().sum())
```

Als nächstes werden diese NaN-Werte durch den String _',No Interpreten,'_ ersetzt, und es wird Kontrolliert ob es keine NaN-Werte mehr gibt:

```python
grammy_relevant_dirty['artist2'] = grammy_relevant_dirty['artist2'].fillna(',No Interpreten,')
grammy_relevant_dirty['Künstler/in'] = grammy_relevant_dirty['Künstler/in'].fillna(',No Interpreten,')

print("Anzahl NaN in 'artist2' nachher: ", grammy_relevant_dirty['artist2'].isna().sum())
print("Anzahl NaN in 'Künstler/in' nachher: ", grammy_relevant_dirty['Künstler/in'].isna().sum())
```

Anschliessend werden die beiden Spalten _'Künstler/in'_ und '_artist2_' zu einer neuen Spalte 'Künstler/in umfassend' zusammengelegt. Dies geschieht einfach mit dem + Operator.

Wichtig zu beachten ist, dass die neue Spalte _'Künstler/in umfassend'_ mit einem "_" gekennzeichnet ist, damit Andere Betrachter sofort sehen, dass es sich um eine angereichtere Spalte handelt:

```python
grammy_relevant_dirty['_Künstler/in umfassend'] = grammy_relevant_dirty['Künstler/in'] + grammy_relevant_dirty['artist2']

grammy_relevant_dirty
```

Bei der Kontrolle der angereicherten Spalte kann nun festgestellt werden, dass immer am Ende des Strings der Zusatz ',No Interpreten,' angefügt wurde. Dies geschieht durch den '+' Operator. Dieser Zusatz muss nun noch in jedem String entfernt werden.

Auch dies wird mit einem Regex-Befehl in der _.replace()_-Methode gemacht. Der Regex:

**(,\w* \w*,)**

sucht nach 2 Wörtern innerhalb von zwei Kommas. Die gefundenen Teil-Strings werden dann durch _''_ (nichts) ersetzt.

```python
grammy_relevant_dirty['_Künstler/in umfassend'].replace('(,\w* \w*,)', '', regex=True, inplace= True)
grammy_relevant_dirty[0:50]
```
